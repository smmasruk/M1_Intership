```python
import cupy
# make an np arrary r=6 to 14 WITH 201 points
r = np.linspace(6.15, 14, 201)
M = 4e6
mu = 5e-2

# make a numpy array named snrProfile
snrProfile = []


e0 = 0.0
theta = np.pi/2  # polar viewing angle
phi =0 # azimuthal viewing angle
dt = 10
T=2000/(365.25*24*3600)
# create a loop for p0 in the range of 6 to 14
for p0 in r:
    wave = few(M, mu, p0, e0, theta, phi, dt=dt, T=2000/(365.25*24*3600))
    # Calculate the Fourier transform of wave.real and wave.imag
    fft_real = np.fft.fft(wave.real)
    fft_imag = np.fft.fft(wave.imag)

    # Calculate the power spectra by squaring the magnitudes
    power_spectrum_real = np.abs(fft_real) ** 2
    power_spectrum_imag = np.abs(fft_imag) ** 2

    # Add the power spectra together
    power_spectrum = power_spectrum_real + power_spectrum_imag
    # size of the power spectrum
    # print(np.shape(power_spectrum))
    # Normalize the power spectrum by dividing it by s_n (PSD)
    sizeofpower = np.shape(power_spectrum)
    # print(sizeofpower)
    normalized_spectrum = power_spectrum 

    # Calculate the SNR by taking the square root of the normalized spectrum
    snr = np.sum(2* np.sqrt(normalized_spectrum))
    # store this snr value in the snrProfile list
    snrProfile.append(snr)

r_norm = r / M

# Calculate np.sqrt(1/T)*snrProfile/mu
sqrt_T = np.sqrt(1 / T)
snr_norm = sqrt_T * np.array(cupy.array(snrProfile).get()) / mu
```

```python
import numpy as np
import cupy as cp
import matplotlib.pyplot as plt

# Define parameters
M = 4e6
mu = 1
dist1 = 0.0000078705
e0 = 0.0
theta = np.pi / 2  # polar viewing angle
phi = 0  # azimuthal viewing angle
dt = 15
r = cp.linspace(10, 16, 201)

# Create a list to store the SNR profile
snrProfile = []

# Define the function to simulate the gravitational wave and calculate SNR
def calculate_snr(p0, L):
    wave = few(M, mu, p0, e0, theta, phi, dist=dist1, dt=dt, T=L)
    fft_real = cp.fft.fft(wave.real)
    fft_imag = cp.fft.fft(wave.imag)
    power_spectrum_real = cp.abs(fft_real) ** 2
    power_spectrum_imag = cp.abs(fft_imag) ** 2
    power_spectrum = power_spectrum_real + power_spectrum_imag
    sum_power = cp.sqrt(cp.sum(power_spectrum))
    return sum_power

# Iterate over different values of T
for L in [1 / 365]:
    snrProfile = []
    for p0 in r:
        snrProfile.append(calculate_snr(p0, L))
    plt.figure()
    plt.plot( np.array(cupy.array(r).get())  ,np.array(cupy.array(snrProfile).get()), '.')
    plt.xlabel('r')
    plt.title(f'SNR Profile for T = {L} years')
    plt.show()
```
