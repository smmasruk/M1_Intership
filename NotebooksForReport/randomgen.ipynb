```python
import random
import numpy as np
import matplotlib.pyplot as plt

# Function to generate Gaussian-distributed random numbers using Box-Muller transform
def box_muller_transform():
    # Generate two independent uniform random numbers
    u1 = random.uniform(0, 1)
    u2 = random.uniform(0, 1)
    
    # Apply the Box-Muller transform
    z0 = np.sqrt(-2.0 * np.log(u1)) * np.cos(2.0 * np.pi * u2)
    z1 = np.sqrt(-2.0 * np.log(u1)) * np.sin(2.0 * np.pi * u2)
    
    return z0, z1

# Generate 9 Gaussian-distributed random numbers within range [0.001, 0.999]
def generate_bounded_gaussian_numbers(n, lower_bound, upper_bound):
    random_numbers = np.empty(n)
    for i in range(0, n, 2):
        z0, z1 = box_muller_transform()
        
        # Scale and shift to fit the range [lower_bound, upper_bound]
        z0 = np.clip((z0 + 3) / 6, lower_bound, upper_bound)  # Standard normal has mean 0 and std dev 1
        z1 = np.clip((z1 + 3) / 6, lower_bound, upper_bound)
        
        random_numbers[i] = z0
        if i+1 < n:
            random_numbers[i+1] = z1

    return random_numbers

random_numbers = generate_bounded_gaussian_numbers(100, 0.001, 0.999)
print(random_numbers)

# Plot the generated random numbers to verify the distribution
plt.hist(random_numbers, bins=50, edgecolor='k')
plt.title('Histogram of Bounded Gaussian-Distributed Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
